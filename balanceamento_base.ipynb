### Acertando as colunas - Tratamento Regex

### Colunas que tinham múltipla escolha

Padrão novo das colunas 

dificuldade_infantil e dificuldade_compra:
- Preco;
- Qualidade;
- Crescimento;
- Preco;Qualidade;
- Preco;Crescimento;
- Qualidade;Crescimento;
- Preco;Qualidade;Crescimento;


forma_trocou:
 - Doou;
 - Trocou;
 - Recebeu;
 - Doou;Trocou;
 - Doou;Recebeu;
 - Trocou;Recebeu;
 - Doou;Trocou;Recebeu;

```python
# Dificuldade de comprar roupar infantis

dificuldade_defective = list(df_pesquisa['dificuldade_infantil'].unique())

for i in dificuldade_defective:
    print(i)
    if i.count(';') == 2:
        if 'Preço' in i:
            if 'Qualidade' in i:
                df_pesquisa = df_pesquisa.replace(i, 'Preco;Qualidade;')
            elif 'Crescimento' in i:
                df_pesquisa = df_pesquisa.replace(i, 'Preco;Crescimento;')
        else:
            df_pesquisa = df_pesquisa.replace(i, 'Qualidade;Crescimento;')
    elif i.count(';') == 3:
        df_pesquisa = df_pesquisa.replace(i, 'Preco;Qualidade;Crescimento;')

# Formas que já trocou roupas

forma_trocou_defective = list(df_pesquisa['forma_trocou'].unique())

for i in forma_trocou_defective:
    print(i)
    if i.count(';') == 2:
        if 'Doou' in i:
            if 'Trocou' in i:
                df_pesquisa = df_pesquisa.replace(i, 'Doou;Trocou;')
            elif 'Recebeu' in i:
                df_pesquisa = df_pesquisa.replace(i, 'Doou;Recebeu;')
        else:
            df_pesquisa = df_pesquisa.replace(i, 'Trocou;Recebeu;')
    elif i.count(';') == 3:
        df_pesquisa = df_pesquisa.replace(i, 'Doou;Trocou;Recebeu;')
```

### Consertando a coluna de cidade

```python
# Função para substituir a cidade
def substituir_cidade(nome):
    cidade = nome.lower().strip()
    if 'guaruj' in cidade:
        return 'Guaruja'
    elif 'guarulhos' in cidade:
        return 'Guarulhos'
    elif 'osasco' in cidade:
        return 'Osasco'
    elif 'barueri' in cidade:
        return 'Barueri'
    elif 'campinas' in cidade or 'camopinas' in cidade:
        return 'Campinas'
    elif 'carapicu' in cidade:
        return 'Carapicuíba'
    elif 'tabo' in cidade:
        return 'Taboao da Serra'
    elif 's' in cidade and 'paulo' in cidade:
        return 'Sao Paulo'
    elif 'sp' in cidade:
        return 'Sao Paulo'
    elif 'saint' in cidade:
        return 'Sao Paulo'
    elif 'cotia' in cidade:
        return 'Cotia'
    else:
        return nome

for i in list(df_pesquisa['cidade'].unique()):
    df_pesquisa = df_pesquisa.replace(i, substituir_cidade(i))

for item, keys in dict(df_pesquisa['cidade'].value_counts()).items():
    print(f'{item}: {keys}')
```

## Preparando para aplicar balanceamento

```python
display(df_pesquisa.head(1))
```

### Separação de X e Y (Para se precisar)

```python
atributos = df_pesquisa.iloc[:, :-1]
resposta = pd.DataFrame(df_pesquisa.iloc[:, -1])
```

### Transformando dados categóricos em númericos

```python
label_encoder = LabelEncoder()

transformador = ColumnTransformer(
    transformers=[('X', OrdinalEncoder(), make_column_selector(dtype_include=['object', 'bool']))],  
    remainder= 'passthrough', 
    verbose_feature_names_out= False
)

df_transformado = pd.DataFrame(transformador.fit_transform(atributos), columns=transformador.get_feature_names_out())

X = df_transformado.sort_index(axis=1)
display(X)

y = pd.DataFrame(label_encoder.fit_transform(resposta), columns= ['Target'])
display(y)

df_transformado_final = pd.concat([X, y], axis=1)
display(df_transformado_final)
```

## Balanceamento das Classes e aumento da base

### SMOTE

```python
smote = SMOTE(sampling_strategy={
        0 : 1037,
        1 : 1001
    },random_state=24)

X_resampled, y_resampled = smote.fit_resample(X, y)

df_smote = pd.concat([X_resampled, y_resampled], axis=1)

display(len(df_smote))
```

### BOOTSTRAP

```python
df_bootstrap = df_transformado_final.sample(n=2038, replace=True, random_state=42)


display(df_bootstrap)
```

### SVM SMOTE

```python
sm = SVMSMOTE(
    sampling_strategy={
        0 : 1037,
        1 : 1001
        }, 
    random_state=42
)

X_res, y_res = sm.fit_resample(X, y)

df_svm_smote = pd.concat([X_res, y_res], axis=1)
display(len(df_svm_smote))
```

### K MEANS SMOTE

```python
k_means = KMeansSMOTE(
    sampling_strategy={
        0 : 1037,
        1 : 1001
    },
    kmeans_estimator=MiniBatchKMeans(n_init=1, random_state=0), 
    random_state=42
)
X_res, y_res = sm.fit_resample(X, y)

df_kmeans_smote = pd.concat([X_res, y_res], axis=1)
display(len(df_kmeans_smote))
```

### BODERLINE SMOTE

```python
boderline = BorderlineSMOTE(
    sampling_strategy={
        0 : 1037,
        1 : 1001
        }, 
        random_state=42)
X_res, y_res = boderline.fit_resample(X, y)

df_boderline_smote = pd.concat([X_res, y_res], axis=1)
display(len(df_boderline_smote))
```

### RANDOM OVERSAMPLING

```python
ros = RandomOverSampler(sampling_strategy={
        0 : 1037,
        1 : 1001
        }, 
        random_state=42)
X_res, y_res = ros.fit_resample(X, y)

df_random_oversampling = pd.concat([X_res, y_res], axis=1)
display(len(df_random_oversampling))
```

### ADASYN

```python
ada = ADASYN(
    sampling_strategy={
        0 : 1037,
        1 : 1001
        },
    random_state=42
    )
X_res, y_res = ada.fit_resample(X, y)

df_adasyn = pd.concat([X_res, y_res], axis=1)
display(len(df_adasyn))
```

# Encapsulando todas as bases geradas

```python
pickle.dump(
    {
     'df_transformado': df_transformado_final,
     'df_smote' : df_smote, 
     'df_bootstrap': df_bootstrap, 
     'df_svm_smote': df_svm_smote, 
     'df_kmeans_smote' : df_kmeans_smote, 
     'df_boderline_smote' : df_boderline_smote, 
     'df_random_oversampling' : df_random_oversampling, 
     'df_adasyn' : df_adasyn
     }, 
    open('./bases/bases.pk1', 'wb'))
```
